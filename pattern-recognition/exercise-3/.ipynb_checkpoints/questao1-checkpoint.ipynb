```python
#agora utilizando o modelo RBF
#importão para gerar os valores de C
from numpy import random
#função para gerar os valores de C e fazer a conversão de tensor para nparray
def generate_c():
    randon_index_lines = np.random.randint(N ,size=int(0.2*N))
    values = x_train[randon_index_lines]
    return torch.from_numpy(np.asarray(values))
    
c = generate_c()
#fazendo a transformação da matriz da distancia entre os cada valor do meu dataset e dos meus c's
Z = torch.cdist(x_train, c)
Z = Z**2

#gerando valor para game, quando menor "melhor"
gamma = 1e-4
cores = torch.exp(-gamma*Z)

#adicionando o bias
Z_train_with_ones = torch.from_numpy(np.concatenate((bias, cores), axis=1))
print(cores.shape, Z_train_with_ones.shape)
```

```python
#calculando o theta do modelo RBF
theta_RBF = torch.inverse(Z_train_with_ones.T @ Z_train_with_ones) @ Z_train_with_ones.T @ y_train
theta_RBF.shape
```

```python
#criando os dados de teste do modelo, fazendo o mesmo tratamento dos dados de treino
Z_test = torch.cdist(torch.from_numpy(x_test), c)
Z_test = Z_test**2
cores_test = torch.exp(-gamma*Z_test)
#adicionando o bias
Z_test_with_ones = torch.from_numpy(np.concatenate((bias_test,Z_test), axis=1))
```
